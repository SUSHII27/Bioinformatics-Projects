```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_excel("/content/covid19_drugs_dataset.xlsx")
```

```python
numeric=df.select_dtypes(include="number")
```

```python
numeric.describe().T
```

```python
categorical=df.select_dtypes(include="object")

categorical.describe().T
```

```python
feature_quality=pd.DataFrame({

"Column":df.columns,

"Missing":df.isnull().sum(),

"Completeness":round(df.notnull().mean()*100,2),

"Unique":df.nunique()

})

feature_quality.head()
```

```python
feature_quality.sort_values(

"Completeness",

ascending=False

).head(30)
```

```python
corr=numeric.corr()
```

```python
fig=px.imshow(

corr,

color_continuous_scale="RdBu_r",

title="Correlation Heatmap"

)

fig.show()
```

```python
import os
os.makedirs('results', exist_ok=True)
feature_quality.to_csv("results/feature_quality.csv")
```
