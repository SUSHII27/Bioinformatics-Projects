```python
import pandas as pd

import plotly.express as px

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA

from sklearn.cluster import KMeans
```

```python
df=pd.read_excel("/content/covid19_drugs_dataset.xlsx")
```

```python
numeric=df.select_dtypes(include="number")
```

```python
numeric=numeric.dropna(

axis=1,

thresh=500
)
```

```python
numeric=numeric.fillna(

numeric.median()

)
```

```python
import numpy as np

# Replace infinity values with NaN
numeric.replace([np.inf, -np.inf], np.nan, inplace=True)

# Re-apply median imputation to handle any newly introduced NaNs from infinity replacement
numeric = numeric.fillna(numeric.median())

scaler=StandardScaler()
scaled=scaler.fit_transform(numeric)
```

```python
pca=PCA(n_components=2)

principal=pca.fit_transform(scaled)

pca_df=pd.DataFrame(

principal,

columns=["PC1","PC2"]

)
```

```python
fig=px.scatter(

pca_df,

x="PC1",

y="PC2",

title="Principal Component Analysis"

)

fig.show()
```

```python
loadings=pd.DataFrame(

pca.components_.T,

columns=["PC1","PC2"],

index=numeric.columns

)

loadings.head()
```

```python
kmeans=KMeans(

n_clusters=5,

random_state=42

)

clusters=kmeans.fit_predict(scaled)

pca_df["Cluster"]=clusters
```

```python
fig=px.scatter(

pca_df,

x="PC1",

y="PC2",

color="Cluster",

title="Compound Clusters"

)

fig.show()
```

```python
import os

# Create the 'results' directory if it doesn't exist
os.makedirs('results', exist_ok=True)

loadings.to_csv(

"results/pca_feature_importance.csv"

)

numeric.to_csv(

"results/processed_dataset.csv",

index=False

)
```
